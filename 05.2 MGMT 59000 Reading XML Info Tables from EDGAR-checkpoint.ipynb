Put it together, we can obtain all table contents from xml links included in Files_Directory/Edgar/Mutual_Fund_Info_Table_Link.txt.

```python
mutual_fund_name = []
info_date = []
name_of_issuer = []
title_of_class = []
CUSIP = []
Value = []
Shares = []
SHRS_or_PRN = []
SH_PRN = []
PUTorCALL = []
Inv_Discretion = []
Other_Manager = []
Authority_Sole = []
Authority_Shared = []
Authority_None = []

CIK = []
xml_date = []
xml_Name = []
xml_link = []

input_file = open('Files_Directory/Edgar/Mutual_Fund_Info_Table_Link.txt', 'r')

rows = input_file.readlines()
input_file.close()

for i in range(1, len(rows)):
    columns = rows[i].split('\t')
    
    CIK.append(columns[0])
    
    if columns[4][:-1] == "xml":
        xml_link.append('https://www.sec.gov' + columns[3])
        xml_Name.append(columns[1])
        xml_date.append(columns[2])


for link_number in range(0,len(xml_link)):
    headers = random.choice(headers_list)
    r = requests.Session()
    r.headers = headers
    html = r.get(xml_link[link_number]).text
    
    soup = bs.BeautifulSoup(html)
    
    infotable = soup.find_all("table", summary = "Form 13F-NT Header Information")
    #get column names
    column_first=infotable[0].find_all('tr')[1].find_all('td')
    column_name_first=[]
    for i in column_first:
        column_name_first.append(i.getText())
    column_name_first=column_name_first+[column_name_first[-1],column_name_first[-1]]
    column_second=infotable[0].find_all('tr')[2].find_all('td')
    column_name_second=[]
    for i in column_second:
        column_name_second.append(i.getText())
    column_name=[a + b for a,b in zip(column_name_first,column_name_second)]
    
    #extract contents
    for row in infotable[0].find_all('tr')[3:]:
        columns = row.find_all('td')
        mutual_fund_name.append(xml_Name[link_number])
        info_date.append(xml_date[link_number])
               
        name_of_issuer.append(columns[column_name.index('NAME OF ISSUER')].getText())
        title_of_class.append(columns[column_name.index('TITLE OF CLASS')].getText())
        CUSIP.append(columns[column_name.index('CUSIP')].getText())
        try:
            Value.append(columns[column_name.index("VALUE(to the nearest dollar)")].getText()+' dollar')
        except:
            Value.append(columns[column_name.index("VALUE(x$1000)")].getText()+' x$1000')
        SHRS_or_PRN.append(columns[column_name.index('SHRS ORPRN AMT')].getText())
        SH_PRN.append(columns[column_name.index('SH/PRN')].getText())
        PUTorCALL.append(columns[column_name.index('PUT/CALL')].getText())
        Inv_Discretion.append(columns[column_name.index('INVESTMENTDISCRETION')].getText())
        Other_Manager.append(columns[column_name.index('OTHERMANAGER')].getText())
        Authority_Sole.append(columns[column_name.index('VOTING AUTHORITYSOLE')].getText())
        Authority_Shared.append(columns[column_name.index('VOTING AUTHORITYSHARED')].getText())
        Authority_None.append(columns[column_name.index('VOTING AUTHORITYNONE')].getText())
        
    time.sleep(random.uniform(5,12))
```
