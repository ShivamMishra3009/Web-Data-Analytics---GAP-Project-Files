```python
websites =['https://www.bestbuy.com/site/apple-iphone-14-pro-max-128gb-deep-purple-verizon/6487406.p?skuId=6487406', 'https://www.bestbuy.com/site/macbook-air-13-6-laptop-apple-m2-chip-8gb-memory-256gb-ssd-silver/6509646.p?skuId=6509646']
prod = []
user = []
rvw = []
count = 1

for link in websites:
    #everything before .p?
    index = link.find('.p?')
    url = link[:index].replace('www.bestbuy.com/site','www.bestbuy.com/site/reviews')
    print(url)#checking url


    headers = random.choice(headers_list)
    r = requests.Session()
    r.headers = headers
    html = r.get(url).text

     # time.sleep(5) # sleep time is 5 seconds
    time.sleep(np.random.uniform(5,15))  #sleep time follows uniform distribution [5,15] with mean is 10
    html = r.get(url).text

    prod_index = html.find('data-track=\"Product Description\"')
    html2 = html[prod_index:]
    start = html2.find(">")
    end = html2.find("<")
    prod_name = html2[start+1:end]

    while html2.find('class=\"review-item\" tabindex=\"-1\"') != -1:

        #prepare prod list
        prod.append(prod_name)

        #prepare user list
        user_index = html2.find('<strong>')
        remaining = html2[user_index+8:]
        end = remaining.find("<")
        user.append(remaining[:end])
        remaining = remaining[end:]

        #prepare review text list
        rvw_index = remaining.find('class="pre-white-space"')
        remaining = remaining[rvw_index:]
        start = remaining.find(">")
        end = remaining.find("<")
        rvw.append(remaining[start+1:end])
        html2 = remaining[end:]
        
    import csv
    # naming files
    file_name = "K_Sathwik_hw4_output"+str(count)+".csv"

    # Combine the lists
    rows = zip(prod, user, rvw)

    # opening and writing in file
    with open(file_name, mode='w', newline='') as file:
        # Create a CSV writer object
        writer = csv.writer(file)

        # Write the header row (optional)
        header = ["Product name", "Reviewer name", "Review" ]
        writer.writerow(header)

        # Write the data rows
        writer.writerows(rows)
        
        print(f"CSV file '{file_name}' is created.")
    #getting ready for next iteration
    count = count+1
    prod = []
    user = []
    rvw = []
```
